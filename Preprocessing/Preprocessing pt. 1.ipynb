# Preprocessing data

### `df1`

```python
# List column names
list(df1.columns)
```

```python
# Selecting relevant columns
df1 = df1[['title', 'protein', 'fat', 'calories']]

# Rename columns
df1.columns = ['Food', 'Protein', 'Fats', 'Energy']
df1.head()
```

```python
# Missing data
df1.isna().sum()
```

```python
# Remove missing data
df1 = df1.dropna()
```

```python
# Duplicate data
df1.duplicated().sum()
```

```python
# Remove duplicate data
df1.drop_duplicates(inplace = True)
```

```python
# Count Carbohydrate
df1['Carbohydrates'] = (df1['Energy'] - 4*df1['Protein'] - 9*df1['Fats'])/4
```

```python
# Change order of columns
df1 = df1.reindex(columns = ['Food', 'Carbohydrates', 'Protein', 'Fats', 'Energy'])
```

```python
# Add extra column Source
df1['Source'] = 'df1'
```

```python
df1.describe()
```

```python
# Remove rows with abnormal values
df1.drop(df1[df1['Energy'] > 700].index, inplace = True)
df1.drop(df1[df1['Energy'] == 0].index, inplace = True)

df1.drop(df1[df1['Carbohydrates'] < 10].index, inplace = True)
df1.drop(df1[df1['Protein'] < 10].index, inplace = True)
df1.drop(df1[df1['Fats'] < 10].index, inplace = True)
df1.describe()
```

```python
df1.shape
```

```python
df1.head()
```

### `df3`

```python
# List column names
list(df3.columns)
```

```python
# Selecting relevant columns
df3 = df3[['fruit', 'carbohydrates (g)', 'protein (g)', 'total fat (g)', 'energy (kcal/kJ)']]

# Rename columns
df3.columns = ['Food', 'Carbohydrates', 'Protein', 'Fats', 'Energy']
df3.head()
```

```python
# Missing data
df3.isna().sum()
```

```python
# Duplicate data
df3.duplicated().sum()
```

```python
# Selecting Energy with kcal units only
df3['Energy'] = df3['Energy'].str.extract(r'(\d+)/\d+').astype(int)
```

```python
# Add extra column Source
df3['Source'] = 'df3'
df3.head()
```

```python
df3.describe()
```

## Integrate datasets and preprocess again

```python
# Combine with df2 and df4
df = pd.concat([df1, df2, df3], axis = 0)
```

```python
# Set first letters of each word in Food to be capitalized
df['Food'] = df['Food'].str.title()
```

```python
# Convert data types
df['Food'] = df['Food'].astype(str)
df['Energy'] = df['Energy'].astype(float)
```
