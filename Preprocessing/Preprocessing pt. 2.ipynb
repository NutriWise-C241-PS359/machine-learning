### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
```

## Importing dataset

```python
# Import df.xlsx
uploaded = files.upload()
```

```python
df = pd.read_excel(io.BytesIO(uploaded['df.xlsx']), sheet_name = 'Clean')
df = pd.DataFrame(df)
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0', 'Source'])
df.head()
```

```python
df.shape
```

## Preprocessing

```python
# Duplicate food names
df['Food'].duplicated().sum()
```

```python
# Remove rows with duplicate food names
df = df.sort_values('Energy', ascending = False).drop_duplicates(subset = 'Food', keep = 'last')
```

```python
df = df.reset_index(drop = True)
df.head()
```

```python
df.shape
```

## Convert data to CSV

```python
# Convert to XLSX
df.to_csv('df_ready.csv')
```
